# Funciones

```python

#Filtro = 

def calculos(Filtro, nombre_columna, variables):
    """
    Realiza diversos cálculos y análisis sobre los DataFrames proporcionados.
    """
    retirados, auxilios, prestamos, empleado_noviembre_2023_activo, empleado_diciembre_2023_consolidado, empleado_diciembre_2023_activo, rotacion_interna_acomulado, finalizacio_curso, solicitudes, vacantes_totales, basse_data, enfermedades_colaborales, sindicato = variables

    # Obtener el tamaño de la gerencia
    tamaño_gerencia = empleado_diciembre_2023_activo.groupby(nombre_columna).size()[Filtro]

    # Conexión con auxilios
    
    # Filtrar empleados que iniciaron antes de 2023
    empleado_diciembre_2023_activo['Fecha de Ingreso'] = pd.to_datetime(empleado_diciembre_2023_activo['Fecha de Ingreso'], format='%d/%m/%Y')
    result = empleado_diciembre_2023_activo[empleado_diciembre_2023_activo['Fecha de Ingreso'].dt.year < 2023]
    conex_auxilio = pd.merge(result, auxilios, on='ID')
    print(conex_auxilio.shape)
    grupo = conex_auxilio.groupby(nombre_columna)
    auxi = calculate_auxilios(grupo, Filtro, tamaño_gerencia)

    # Conexión con préstamos
    conex_prestamos = pd.merge(empleado_diciembre_2023_activo, prestamos, on='ID')
    grupo = conex_prestamos.groupby(nombre_columna)
    pres = calculate_prestamos(grupo, Filtro)

    # Índice de rotación
    indice_rot = calculate_rotacion(empleado_diciembre_2023_activo, retirados, empleado_noviembre_2023_activo, nombre_columna, Filtro, tamaño_gerencia)

    # Finalización de cursos
    porcentaje_cursos = calculate_cursos(empleado_diciembre_2023_activo, finalizacio_curso, nombre_columna, Filtro)

    # Base data
    sanas, promedio = calculate_productividad(basse_data, empleado_diciembre_2023_activo, nombre_columna, Filtro, tamaño_gerencia)

    # Enfermedades colaborales
    base_colab = calculate_enfermedades_colaborales(empleado_diciembre_2023_activo, enfermedades_colaborales, nombre_columna, Filtro)

    # Sindicato
    no_pertenece = calculate_sindicato(empleado_diciembre_2023_activo, sindicato, nombre_columna, Filtro, tamaño_gerencia)

    return [no_pertenece, sanas, promedio, porcentaje_cursos, auxi, indice_rot]

# Funciones auxiliares para cálculos específicos
def calculate_auxilios(grupo, Filtro, tamaño_gerencia):
    """
    Calcula el porcentaje de empleados con auxilios.
    """
    if Filtro in grupo.groups:
        auxil_data_frame = grupo.get_group(Filtro)
        recuento_uniq = auxil_data_frame['ID'].unique()
        auxi = (recuento_uniq / tamaño_gerencia) * 100
    else:
        auxi = 0
    print(auxil_data_frame.shape)
    return auxi

def calculate_prestamos(grupo, Filtro):
    """
    Calcula el monto total de préstamos.
    """
    if Filtro in grupo.groups:
        pres_data_frame = grupo.get_group(Filtro)
        total_pres = pres_data_frame['Cuota'].sum()
    else:
        total_pres = 0
    return total_pres

def calculate_rotacion(empleado_diciembre_2023_activo, retirados, empleado_noviembre_2023_activo, nombre_columna, Filtro, tamaño_gerencia):
    """
    Calcula el índice de rotación de empleados.
    """
    n = empleado_diciembre_2023_activo.groupby(nombre_columna).size()[Filtro]
    E = retirados.groupby(nombre_columna).size()[Filtro]
    s = empleado_noviembre_2023_activo.groupby(nombre_columna).size()[Filtro]
    indice_rot = (E / (n + s)) * 100
    return indice_rot

def calculate_cursos(empleado_diciembre_2023_activo, finalizacio_curso, nombre_columna, Filtro):
    """
    Calcula el porcentaje de empleados que finalizaron cursos.
    """
    finalizacion_curso_conexo = pd.merge(empleado_diciembre_2023_activo, finalizacio_curso, on='ID')
    grupo = finalizacion_curso_conexo.groupby(nombre_columna)
    if Filtro in grupo.groups:
        data_frame_final = grupo.get_group(Filtro)
        por = data_frame_final.groupby('Nombre del Curso').size() / data_frame_final['ID'].nunique()
        porcentaje_cursos = por.mean()
    else:
        porcentaje_cursos = 0
    return porcentaje_cursos

def calculate_productividad(basse_data, empleado_diciembre_2023_activo, nombre_columna, Filtro, tamaño_gerencia):
    """
    Calcula el número de empleados sin accidentes y el promedio de productividad.

    """
    conex_basedata = pd.merge(empleado_diciembre_2023_activo, basse_data, on='ID')
    grupo_data = conex_basedata.groupby(nombre_columna)

    # Verifica si el valor filtrado está presente en el índice
    if Filtro in grupo_data.groups:
        # Obtén los datos para el valor filtrado
        rot = grupo_data.size()[Filtro]
        na = conex_basedata.query(f"{nombre_columna} == '{Filtro}'")
        na_nueva = 1
    else:
        na_nueva = 0
        print(f"No hay datos disponibles para {Filtro}.")

    if na_nueva != 0:
        sanas = ((tamaño_gerencia - na.shape[0])/tamaño_gerencia)*100
    else:
        sanas = ((tamaño_gerencia -0)/tamaño_gerencia)*100

    if na_nueva == 0:
        promedio = 0
    else:
        basse_data_copia_sin_na = na.dropna(subset=['% Productividad'])
        filas_filtradas_100 = basse_data_copia_sin_na[basse_data_copia_sin_na['% Productividad'].str.contains('100%')].shape[0]
        filas_filtradas_75 = basse_data_copia_sin_na[basse_data_copia_sin_na['% Productividad'].str.contains('75%')].shape[0]
        filas_filtradas_50 = basse_data_copia_sin_na[basse_data_copia_sin_na['% Productividad'].str.contains('50%')].shape[0]
        filas_filtradas_0 = basse_data_copia_sin_na[basse_data_copia_sin_na['% Productividad'].str.contains('No cumple%')].shape[0]
        promedio = (filas_filtradas_100*100 + filas_filtradas_75*75 + filas_filtradas_50*50 +filas_filtradas_0*0)/(filas_filtradas_100+filas_filtradas_75+filas_filtradas_50+filas_filtradas_0)
    return sanas,promedio

def calculate_enfermedades_colaborales(empleado_diciembre_2023_activo, enfermedades_colaborales, nombre_columna, Filtro):
    """
    Calcula el número de empleados con enfermedades laborales.
    """
    enfermedades_colaborales_conexo = pd.merge(empleado_diciembre_2023_activo, enfermedades_colaborales, on='ID')
    grupo = enfermedades_colaborales_conexo.groupby(nombre_columna)
    if Filtro in grupo.groups:
        base_colab = grupo.get_group(Filtro)['ID'].nunique()
    else:
        base_colab = 0
    return base_colab

def calculate_sindicato(empleado_diciembre_2023_activo, sindicato, nombre_columna, Filtro, tamaño_gerencia):
    """
    Calcula el porcentaje de empleados que no pertenecen al sindicato.
    """
    sindicato_conexo = pd.merge(empleado_diciembre_2023_activo, sindicato, on='ID')
    grupo = sindicato_conexo.groupby(nombre_columna)
    if Filtro in grupo.groups:
        grupo_sindicato = grupo.get_group(Filtro)
        num_sindicato = grupo_sindicato['ID'].nunique()
    else:
        num_sindicato = 0
    no_pertenece = (tamaño_gerencia - num_sindicato) / tamaño_gerencia * 100
    return no_pertenece
```

```python


def DashApp():
    opciones_1 = ["Centro  de Trabajo", "Cargo", "Centro Costo Gerentes"]
    
    # Ordenar y asignar las listas al diccionario opciones_2
    opciones_2 = {
        "Centro  de Trabajo": sorted(datos_limpios[5]['Centro  de Trabajo'].value_counts()[lambda x: x > 100].index.tolist()),
        "Cargo": sorted(datos_limpios[5]['Cargo'].value_counts()[lambda x: x > 100].index.tolist()),
        "Centro Costo Gerentes": sorted(datos_limpios[5]['Centro Costo Gerentes'].value_counts()[lambda x: x > 100].index.tolist())
    }

    app = dash.Dash(__name__, suppress_callback_exceptions=True)

    app.layout = html.Div([
        dcc.Tabs(id='tabs', value='tab-1', children=[
            dcc.Tab(label='Gráfico Radial', value='tab-2'),
        ]),
        html.Div(id='tabs-content')
    ])

    @app.callback(
        Output('tabs-content', 'children'),
        Input('tabs', 'value')
    )
    def render_content(tab):
        if tab == 'tab-2':
            return html.Div([
                dcc.Dropdown(
                    id='filtro-dropdown-1',
                    options=[{'label': filtro, 'value': filtro} for filtro in opciones_1],
                    value=opciones_1[0]
                ),
                dcc.Dropdown(id='filtro-dropdown-2'),
                html.Div([
                    dcc.Graph(id='grafico-radial'),
                    html.Div([
                        html.P("- Sindicato: Porcentaje de personas que no hacen parte del sindicato", style={'display': 'block'}),
                        html.P("- Incapacidad: Porcentaje de personas en la nóminca que están incapacitados", style={'display': 'block'}),
                        html.P("- Prod. incapacidad: Porcentaje de productividad cuando están incapacitados", style={'display': 'block'}),
                        html.P("- Educación: Porcentaje de cursos obligatorios que han relaizado", style={'display': 'block'}),
                        html.P("- Auxilios: Porcentaje de personas que han pedido al menos un axilio en el último año", style={'display': 'block'}),
                        html.P("- índice rotación: Porcentaje de personas que no han", style={'display': 'block'})
                    ], style={'margin-top': '20px', 'text-align': 'center', 'line-height': '1.5'})
                ], style={'display': 'flex', 'flex-direction': 'column', 'align-items': 'center'})
            ])
        return html.Div()  # Agregar un valor de retorno para 'tab-1' o contenido predeterminado

    @app.callback(
        Output('filtro-dropdown-2', 'options'),
        Input('filtro-dropdown-1', 'value')
    )
    def update_options(selected_filter):
        if selected_filter:
            return [{'label': val, 'value': val} for val in opciones_2[selected_filter]]
        return []

    @app.callback(
        Output('grafico-radial', 'figure'),
        [Input('filtro-dropdown-1', 'value'), 
         Input('filtro-dropdown-2', 'value')]
    )
    def update_graph(filtro_1, filtro_2):
        if filtro_1 and filtro_2:
            # Suponiendo que 'calculos' es una función definida que procesa los datos
            datos = calculos(filtro_2, filtro_1, datos_limpios)
            print(datos)
            df = pd.DataFrame([datos], columns=["Sindicato", 'Incapacidad', 'Product. Incapacidad', 'Educación', 'Auxilios', 'Índice de rotación'])

            # Personalización del gráfico radial
            fig = px.scatter_polar(df, title=f"{filtro_1} - {filtro_2}", labels={"index": "Categoría"}, 
                                   theta=df.columns, r=df.iloc[0],
                                   color_discrete_sequence=px.colors.sequential.Plasma)  # Cambia la secuencia de colores

            fig.update_traces(fill='toself')

            fig.update_layout(
                width=600,  # Ancho del gráfico
                height=600,  # Alto del gráfico
                polar=dict(
                    radialaxis=dict(
                        visible=True  # Ocultar las etiquetas radiales
                    ),
                    angularaxis=dict(
                        showline=True,
                        linewidth=2,
                        linecolor='black',
                        tickfont=dict(size=16)
                    ),
                    bgcolor="lightblue"  # Cambia el fondo de la zona polar
                ),
                title={
                    'text': f"{filtro_1} - {filtro_2}",
                    'y':0.9,
                    'x':0.5,
                    'xanchor': 'center',
                    'yanchor': 'top'
                },
            )

            return fig
        # Si algún filtro no está seleccionado, devolver un gráfico vacío
        return {'data': [], 'layout': {}}

    if __name__ == '__main__':
        app.run_server(debug=True, port=8052)

DashApp()
```
